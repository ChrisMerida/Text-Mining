### Problema 2

```python
def problema_dos(img1, img2):
    dim_img1 = img1.shape
    dim_img2 = img2.shape

    if dim_img1 == dim_img2:
        print("Mismo tamano")
        
        
    else:
        print("Tamano distinto")
        img2 = cv2.resize(img2, (dim_img1[1], dim_img1[0]))
        
    img1_gris = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
    img2_gris = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
    
    suma_imagenes = cv2.add(img1, img2)
    resta_imagenes = cv2.subtract(img1, img2)

    # Imagen 1 y 2
    plt.figure(figsize=(6, 3))
    plt.subplot(1, 2, 1)
    plt.title("Imagen 1")
    plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Imagen 2")
    plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    # Suma y Resta
    plt.figure(figsize=(6, 3))
    plt.subplot(1, 2, 1)
    plt.title("Suma de Imágenes")
    plt.imshow(cv2.cvtColor(suma_imagenes, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Resta de Imágenes")
    plt.imshow(cv2.cvtColor(resta_imagenes, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # Binarización
    _, img1_binarizada = cv2.threshold(img1_gris, 127, 255, cv2.THRESH_BINARY)
    _, img2_binarizada = cv2.threshold(img2_gris, 127, 255, cv2.THRESH_BINARY)

    # Operaciones lógicas
    bitwise_and = cv2.bitwise_and(img1_binarizada, img2_binarizada)
    bitwise_or = cv2.bitwise_or(img1_binarizada, img2_binarizada)
    bitwise_xor = cv2.bitwise_xor(img1_binarizada, img2_binarizada)

    # Imágenes binarizadas
    plt.figure(figsize=(6, 3))
    plt.subplot(1, 2, 1)
    plt.title("Imagen Binarizada 1")
    plt.imshow(img1_binarizada, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Imagen Binarizada 2")
    plt.imshow(img2_binarizada, cmap='gray')
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(6, 3))
    plt.subplot(1, 3, 1)
    plt.title("AND")
    plt.imshow(bitwise_and, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("OR")
    plt.imshow(bitwise_or, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("XOR")
    plt.imshow(bitwise_xor, cmap='gray')
    plt.axis('off')
    plt.show()
    
```

```python
def problema_tres(imagen):
    if len(imagen.shape) == 3 and imagen.shape[2] == 3:
        imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    alto, ancho = imagen.shape
    
    x = np.arange(0, ancho)
    y = np.arange(0, alto)
    x, y = np.meshgrid(x, y)
    
    z = imagen
    
    # Crear una figura y un conjunto de ejes 3D
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Crear la gráfica 3D
    ax.plot_surface(x, y, z, cmap='gray', edgecolor='none')
    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Intensidad')
    ax.set_title('Visualización 3D de Imagen en Escala de Grises')
    
    plt.show()
```

```python
imagen = cv2.imread("iphone.jpg", cv2.IMREAD_GRAYSCALE)
problema_tres(imagen)
```

```python
def lineal(imagen, alpha, beta): 
    imagen_transformada = cv2.convertScaleAbs(imagen, alpha=alpha, beta=beta)

    plt.figure(figsize=(6, 3))

    plt.subplot(1, 2, 1)
    plt.title("Imagen Original")
    plt.imshow(imagen)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Imagen Transformada")
    plt.imshow(imagen_transformada)
    plt.axis('off')

    plt.show()
```

```python
imagen = cv2.imread("iphone.jpg")
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

alpha = 1.5
beta = 50
```

```python
prueba_cuatro = lineal(imagen, alpha, beta)
```

```python
def logaritmica(imagen, c):
    if len(imagen.shape) == 3 and imagen.shape[2] == 3:
        imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

    imagen_conv = imagen.astype(np.float32)
    imagen_log = c * np.log(1 + imagen_conv)

    imagen_log = np.array(255 * (imagen_log / np.max(imagen_log)), dtype=np.uint8)

    plt.figure(figsize=(6, 3))

    plt.subplot(1, 2, 1)
    plt.title("Imagen Original")
    plt.imshow(imagen)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Imagen Transformada")
    plt.imshow(imagen_log)
    plt.axis('off')

    plt.show()
    
```

```python
imagen = cv2.imread("iphone.jpg")
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

c = 1.0
```

```python
prueba_log  = logaritmica(imagen, c)
```

```python
def exponencial(imagen, c, beta):
    if len(imagen.shape) == 3 and imagen.shape[2] == 3:
        imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

    imagen_conv = imagen.astype(np.float32)
    imagen_exp = c * (np.exp(imagen_conv / beta) - 1)

    imagen_exp = np.array(255 * (imagen_exp / np.max(imagen_exp)), dtype=np.uint8)

    plt.figure(figsize=(6, 3))

    plt.subplot(1, 2, 1)
    plt.title("Imagen Original")
    plt.imshow(imagen, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Imagen Transformada")
    plt.imshow(imagen_exp, cmap='gray')
    plt.axis('off')

    plt.show()
```

```python
imagen = cv2.imread("iphone.jpg")
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

c = 1.0
beta = 30.0
```

```python
prueba_exp = exponencial(imagen, c, beta)
```

### Problema 5

El kernel Laplaceano es un operador de detección de bordes basado en el segundo derivado de imagen. Calcula el Laplaceano de una imagen, el cual es una medida de la tasa de cambio del gradiente en cada pixel. Se utiliza una mascara de convolusión que evalúa la segunda derivada en cada punto de la imagen, identificando los bordes en donde hay una rápida variación en la intensidad de la imagen.   
Este operador detecta bordes en todas las direcciones de la imagen, sin embargo, es sensible al ruido y puede generar bordes discontinuos.   

El algoritmo Canny es un enfoque de detección de bordes más sofisticado. Es necesario seguir una secuencia de pasos empezando por el suavizado de imagen, luego se calcula el gradiente de la imagen, se elimina los píxeles que no son bordes locales y utiliza dos umbrales para determinar bordes fuertes y debiles y conectarlos. Este algoritmo es mas robusto al ruido y produce bordes más finos y continuos, sin embargo, es más complejo y requiere ajustes de parámetros.  

Diferencias  
1. Laplaceano utiliza la segunda derivada para detectar bordes. Canny utiliza una serie de pasos.
2. Canny es más robusto frente al ruido y proporciona bordes más precisos y continuos. Laplaceano es más sensible y menos preciso para la detección de bordes.
3. Canny es más complejo y requiere más parámetros.
