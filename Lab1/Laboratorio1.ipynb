```python
def problema_uno(img, color):
    img = cv2.imread(dir_imagen)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    img_salida = np.zeros_like(img)

# Color azul = 2
# Color verde = 1
# Color rojo = 0

    if color == 1:
        img_salida[:, :, 2] = img[:, :, 2]
        color_activo = "Color Azul Activo"
    
    elif color == 2:
        img_salida[:, :, 1] = img[:, :, 1]
        color_activo = "Color Verde Activo"

    elif color == 3:
        img_salida[:, :, 0] = img[:, :, 0]
        color_activo = "Color Rojo Activo"

    elif color == 10:
        img_salida[:, :, 0] = img[:, :, 0]
        img_salida[:, :, 1] = img[:, :, 1]
        color_activo = "Color Rojo y Verde Activos"

    elif color == 20:
        img_salida[:, :, 1] = img[:, :, 1]
        img_salida[:, :, 2] = img[:, :, 2]
        color_activo = "Color Verde y Azul Activos"

    elif color == 30:
        img_salida[:, :, 0] = img[:, :, 0]
        img_salida[:, :, 2] = img[:, :, 2]
        color_activo = "Color Rojo y Azul Activos"

    else:
        img_salida = img
        color_activo = "Imagen Original"

    plt.figure(figsize=(10, 5))
    plt.imshow(img_salida)
    plt.axis('off')
    plt.title(color_activo)
    plt.show()

    return img_salida
```

### Problema 2

```python
def problema_dos(dir_img1, dir_img2, dir_img3):
    img1 = cv2.imread(dir_img1)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

    img2 = cv2.imread(dir_img2)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

    img3 = cv2.imread(dir_img3)
    img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
    
    canal_rojo = img1[:, :, 0]
    canal_verde = img2[:, :, 1]
    canal_azul = img3[:, :, 2]

    img_original = np.zeros_like(img1)

    img_original[:, :, 0] = canal_rojo
    img_original[:, :, 1] = canal_verde
    img_original[:, :, 2] = canal_azul

    plt.figure(figsize=(10, 5))
    plt.imshow(img_original)
    plt.title("Imagen Original")
    plt.axis('off')
    plt.show()

    return img_original
```

```python
# Imagen 1
dir_img1 = "imagen1/imagen1_salida_gray_rojo.jpg"
dir_img2 = "imagen1/imagen1_salida_gray_verde.jpg"
dir_img3 = "imagen1/imagen1_salida_gray_azul.jpg"
prueba_dos = problema_dos(dir_img1, dir_img2, dir_img3)
```

### Problema 4

```python
def problema_cuatro(dir_img):
    img_color = cv2.imread(dir_img)
    img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

    # Canal Rojo
    histograma_rojo = cv2.calcHist([img_color], [0], None, [256], [0, 256])
    rojo_media = np.mean(img_color[:, :, 0])
    
    plt.figure(figsize=(6, 3))
    plt.plot(histograma_rojo, color='red')
    plt.axvline(rojo_media, color='k', linestyle='dashed', linewidth=1)
    plt.xlabel("Intensidad del Pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma Canal Rojo")
    plt.show()

   # Canal Verde
    histograma_verde = cv2.calcHist([img_color], [1], None, [256], [0, 256])
    verde_media = np.mean(img_color[:, :, 1])
    
    plt.figure(figsize=(6, 3))
    plt.plot(histograma_verde, color='green')
    plt.axvline(verde_media, color='k', linestyle='dashed', linewidth=1)
    plt.xlabel("Intensidad del Pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma Canal Verde")
    plt.show()

    # Canal Azul
    histograma_azul = cv2.calcHist([img_color], [2], None, [256], [0, 256])
    azul_media = np.mean(img_color[:, :, 2])
    
    plt.figure(figsize=(6, 3))
    plt.plot(histograma_azul, color='blue')
    plt.axvline(azul_media, color='k', linestyle='dashed', linewidth=1)
    plt.xlabel("Intensidad del Pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma Canal Azul")
    plt.show()

    # Escala de grises
    imagen_gris = np.mean(img_color, axis=2).astype(np.uint8)
    histograma_gris = cv2.calcHist([imagen_gris], [0], None, [256], [0, 256])
    gris_media = np.mean(imagen_gris)
    
    plt.figure(figsize=(6, 3))
    plt.plot(histograma_gris)
    plt.axvline(gris_media, color='k', linestyle='dashed', linewidth=1)
    plt.xlabel("Intensidad del Pixel")
    plt.ylabel("Cantidad de Pixeles")
    plt.title("Histograma Escala Grises")
    plt.show()
```

```python
# Prueba
dir_img = "prueba.jpg"
prueba_tres = problema_cuatro(dir_img)
```

### Problema 5

Método para convertir imágenes en color a escala de grises donde se toma en cuenta diferentes pesos para cada uno de los canales de coor. El ojo humano no percibe todas las intensidades, aplicar este método permite que la conversión refleje mejor la percepción visual humana de los colores.  

Para imlementar una conversión de escala de grises ponderada, es necesario asignar coeficientes a cada uno de los canales de color para calcular el valor de gris final.  
Comúnmente, se utiliza 0.29 en canal rojo, 0.59 en el canal verde y 0.11 en el canal azul. Estos valores surgen de un estudio sobre la sensibilidad del ojo humano a diferentes colores y se catalogan como estándar.  

La fórmula general para convertir un color RGB a gris es:  
'Utilizando los valores estándar'  
Escala de gris = 0.29xRojo + 0.59xVerde + 0.11xAzul

```python
def problema_cinco(imagen_ponderada):
    imagen_ponderada = cv2.imread(dir_img)
    imagen_ponderada = cv2.cvtColor(imagen_ponderada, cv2.COLOR_BGR2RGB)

    img_gris = cv2.cvtColor(imagen_ponderada, cv2.COLOR_BGR2GRAY)

    img_color_float = imagen_ponderada.astype(float)
    img_gris_manual = 0.2989 * img_color_float[:, :, 2] + 0.5870 * img_color_float[:, :, 1] + 0.1140 * img_color_float[:, :, 0]
    img_gris_manual = np.uint8(img_gris_manual)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img_gris, cmap='gray')
    plt.title("Escala de Grises Ponderada OpenCV")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(img_gris_manual, cmap='gray')
    plt.title("Escala de Grises Ponderada Manual")
    plt.axis('off')
    plt.show()
```

### Problema 6

El espacio de color HSV (Hue, Saturation, Value) es una representación alternativa del espacio de color RGB. Describe los colores de una manera que es más natural para la percepción humana.  
* Hue: representa el tipo de color. Se mide en grados en un círculo los cuales corresponden a rojo, verde y azul. 
* Saturation: indica la pureza de color, variando desde 0% hasta 100%. El valor de 0% representa un color gris, es decir, sin saturación y 100% representa un color completamente puro, máxima saturación. 
* Value: indica la luminosidad del color, variando desde 0% hasta 100. El valor 0% es completamente negro y 100% es el color más brillante posible.  

El espacio de color HSV es más intuitivo debido a que separa la percepción de color (Hue), de la percepción de briollo y pureza (Value y Saturation). Ádemas, es más facil realizar ciertas operaciones de manipulación de imagenes sin modificar otros aspectos de color.  

Conversión  
La conversión de RGB a HSV implica una serie de cálculos que transforman los valores RGB en los componentes HSV. Es necesario identificar el color más intenso y el menos intenso para calcular el tono, la saturación y el valor.  
La conversión de HSV a RGB implica una conversión inversam donde se calculan los valores de RGB a partir de HSV.  

R: Componente roja (0-255)  
B: Componente azul (0-255)  
G: Componente verde (0-255)  

1. Normalizar los datos:   
r = R/255, b = B/255, g = G/255  
2. Valores mínimo, máximo y rango de componentes normalizados.  
Max(R, G, B)  
Min(R, G, B)  
Rango = Max - Min
3. Calcular Hue.
4. Calcular Saturation.
5. Calcular Value.
6. Retornar los valores en una imagen.
